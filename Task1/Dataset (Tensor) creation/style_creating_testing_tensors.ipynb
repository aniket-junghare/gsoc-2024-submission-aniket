```python
style_val_dataset = pd.read_csv("C:/Users/23M0321/Desktop/GSOC - Task 1/CSV Files/wikiart_csv/style_val.csv")
```

```python
val_images = style_val_dataset['Image']
val_style = style_val_dataset['Style']
```

```python
tensor_list = []
```

```python
final_val_styles = []
```

```python
width = 224
height = 224
```

```python
from PIL import Image
import torchvision.transforms as transforms
import torch
```

```python
transform = transforms.Compose([
    transforms.Resize((width, height)),  
    transforms.ToTensor() 
])
```

```python
for i in range(0,24421,3):
    folder_name , filename = val_images[i].split("/")
    image_path = r"C:\Users\23M0321\Desktop\GSOC - Task 1\wikiart\wikiart\{}\{}".format(folder_name, filename)
    if not os.path.exists(image_path):
        print(f"File {image_path} does not exist. Skipping...")
        continue
    image = Image.open(image_path)
    resized_image = transform(image)
    tensor_list.append(resized_image)
    final_val_styles.append(val_style[i])
    print(i)
```
